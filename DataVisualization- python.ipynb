```python
def slice_unrate(year):
    return unrate[unrate["Year"]==year]
unrate_1948= slice_unrate(1948)
unrate_1949 = slice_unrate(1949)
unrate_1950 = slice_unrate(1950)
unrate_1951 = slice_unrate(1951)
unrate_1952 = slice_unrate(1952)

fig = plt.figure(figsize=(12,12))

ax1 = fig.add_subplot(5,1,1)
ax2 = fig.add_subplot(5,1,2)
ax3 = fig.add_subplot(5,1,3)
ax4 = fig.add_subplot(5,1,4)
ax5 = fig.add_subplot(5,1,5)

ax1.plot(unrate_1948["DATE"],unrate_1948["VALUE"])
ax2.plot(unrate_1949["DATE"],unrate_1949["VALUE"])
ax3.plot(unrate_1950["DATE"],unrate_1950["VALUE"])
ax4.plot(unrate_1951["DATE"],unrate_1951["VALUE"])
ax5.plot(unrate_1952["DATE"],unrate_1952["VALUE"])


plt.show()
```

```python
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax = fig.add_subplot(5,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])

plt.show()
```

```python
unrate['MONTH'] = unrate['DATE'].dt.month

fig = plt.figure(figsize=(6,3))
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='blue')

plt.show()
```

```python
color =['red','blue','green','orange','black']

fig = plt.figure(figsize=(10,6))
    
for i,value in enumerate(color):
    plt.plot(unrate[12*i:12*(i+1)]['MONTH'], unrate[12*i:12*(i+1)]['VALUE'], c=value)
plt.show()
```

```python
color =['red','blue','green','orange','black']

fig = plt.figure(figsize=(10,6))
    
for i,value in enumerate(color):
    plt.plot(unrate[12*i:12*(i+1)]['MONTH'], unrate[12*i:12*(i+1)]['VALUE'], c=value, label=str(1948+i))
plt.legend(loc='upper left')
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    label = str(1948 + i)
    plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=label)
plt.legend(loc='upper left')
plt.title("Monthly Unemployment Trends, 1948-1952")
plt.xlabel("Month, Integer")
plt.ylabel("Unemployment Rate, Percent")
plt.show()
```
